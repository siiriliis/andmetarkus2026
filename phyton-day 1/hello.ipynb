```python
year = -200000
if year < 0:
    print("B.C")
    if year < -10000:
        print("Stone age..")
elif year == 0:
    print("Happy birthday, Jesus!")
elif year == 1978:
    print("Happy Birthday, Mati!")
else:
    print("A.D.")
```

```python
def make_sense_of_this_year(year: int):
    if year < 0:
        print("B.C")
        if year < -10000:
            print("Stone age..")
    elif year == 0:
        print("Happy birthday, Jesus!")
    elif year == 1978:
        print("Happy Birthday, Mati!")
    else:
        print("A.D.")

make_sense_of_this_year(-90)
make_sense_of_this_year(year=-90)
```

```python
def print_year(year: int):
    if year < 0:
        print(f"{abs(year)} B.C")
    else:
        print(f"{abs(year)} A.D")

print_year(-90)
print_year(2025)
```

```python
from sklearn.datasets import load_iris
import pandas as pd

iris_data = load_iris()
iris = pd.DataFrame(iris_data.data, columns = iris_data.feature_names)
iris['target'] = iris_data.target
iris['species'] = iris['target'].map(dict(enumerate(iris_data.target_names)))
iris.pop('target')
iris = iris.rename(columns={
    "sepal length (cm)" : "sepal_length",
    "sepal width (cm)" : "sepal width",
    "petal length (cm)" : "petal_length",
    "petal width (cm)" : "petal width"
})
iris
```

```python
iris.to_csv("iris.csv", index=False)
```

## HARJUTUS DAY 2

- vaja leida kogu hind
- kui hind on üle 100, siis 20% alla
- kui hind on üle 50, siis 10% alla
- lõpptulemuseks on funktsioon get.price(hinnad)

```python
# minu lahendus
hinnad = [120,45,89,200,33,75,150,10,99,101]

def get_price(list_sisendiks: int) ->float:
    uued_hinnad = []
    for hind in list_sisendiks:
        if hind > 100:
            uus_hind = hind*0.8
            uued_hinnad.append(uus_hind)
            
        elif hind > 50:
            uus_hind = hind*0.9
            uued_hinnad.append(uus_hind)

        else:
            uued_hinnad.append(hind)

    return sum(uued_hinnad)
lopphind = get_price(hinnad)

print(f"Kogu hind on: {lopphind}€ ")  

```

```python
# õpetaja lahendus
hinnad = [120,45,89,200,33,75,150,10,99,101]

def get_discount(price: int) -> float:
        if price > 100:
            return price*0.8
            
        if price > 50:
            return price*0.9
        return price

def get_price(uued_hinnad: list[int]) -> float:
     tulemus = 0
     for uus_hind in uued_hinnad:
          allahindlusega_hind = get_discount(uus_hind)
          tulemus = tulemus + allahindlusega_hind
     return tulemus
     
    
get_price(hinnad)
```

```python
# õpetaja lahendus 2
hinnad = [120,45,89,200,33,75,150,10,99,101]

def get_discount(price: int) -> float:
        if price > 100:
            return price*0.8
            
        if price > 50:
            return price*0.9
        return price

def get_price_short(uued_hinnad: list[int]) -> float:
     return sum([get_discount(hind) for hind in uued_hinnad])

    
get_price_short(hinnad)
```
