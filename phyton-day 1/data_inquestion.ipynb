```python
def numeric_year(year_str: str) -> int:
    year_number_components: list[str] = year_str.split(" ")
    if len(year_number_components) != 2:
        raise ValueError(f" '{year_str}' is not a valid year number: missing space")
    year_number_str: str = year_number_components[0]
    year_era_notation_str: str = year_number_components[1]
   
    try:  
        year_numeric = int(year_number_str)
    except:
        raise ValueError(f" '{year_str}' is not a valid year number: non-numeric year number")

    if year_era_notation_str == "A.D.":
        return year_numeric
    elif year_era_notation_str == "B.C.":
        return -year_numeric
    else:
        raise ValueError(f" '{year_str}' is not a valid year number: unknown era ")


print(numeric_year("2024 A.D.")) #> 2024
print(numeric_year("1 B.C.")) #> -1
print(numeric_year("12 A.D.")) # 12
print(numeric_year("kaksteist A.D.")) #> ValueError: 'kaksteist AD.' does not seem to be a valid year number

```
